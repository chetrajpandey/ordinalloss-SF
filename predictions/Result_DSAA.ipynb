```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np

def round_up(x, decimals=0):
    multiplier = 10 ** decimals
    result = math.ceil(x * multiplier) / multiplier
    result = format(result, f".{decimals}f")
    return result

# Increase the font size of all elements
plt.rcParams.update({'font.size': 15})

# Assuming the data is in a CSV file named 'data.csv'
data = pd.read_csv('overall_Result.csv')

# Separate the data based on the model
bce = data[data['Loss Function'] == 'bce']
bce_sf = data[data['Loss Function'] == 'bcesf']

# Define the lon_fwt Range categories
categories = ["(-30, 30)", "(-60, 60)", "(-90, 90)"]

# Define the metrics
metrics = ['TSS (Validation)', 'HSS (Validation)', 'Geomean (Validation)']

# Define colorblind-friendly colors
colors = ['#56B4E9', '#009E73']
ylabel = ['True Skill Statistic (TSS)', 'Hiedke Skill Score (HSS)', 'Composite Skill Score (CSS)']

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
axes = axes.flatten()
bar_width = 0.35
# For each metric, create a bar plot
for j, (metric, ax) in enumerate(zip(metrics, axes)):
    # For each category, get the metric value for each model and plot it
    for i, category in enumerate(categories):
        bce_value = bce[bce['lon_fwt Range'] == category][metric].values[0]
        bcesf_value = bce_sf[bce_sf['lon_fwt Range'] == category][metric].values[0]
        
        ax.bar(i - bar_width/2, bce_value, bar_width, color=colors[0], align='center')
        ax.bar(i + bar_width/2, bcesf_value, bar_width, color=colors[1], align='center')
        
        # Add the values on top of each bar
        ax.text(i - bar_width/2, bce_value, round_up(bce_value, 2), ha='center', va='bottom')
        ax.text(i + bar_width/2, bcesf_value, round_up(bcesf_value, 2), ha='center', va='bottom')
    
    # Add labels and legend
    ax.set_xticks(range(len(categories)))
    ax.set_xticklabels(["±30°", "±60°", "±90°"])
    ax.set_xlabel('Solar Longitude in degrees (°)')
    ax.set_ylabel(ylabel[j])
    ax.legend(['BCE', 'BCE-SF'], loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=2, frameon=False)
    
    if metric == 'TSS (Validation)':
        ax.set_yticks(np.arange(0, 0.8, 0.2))
    else: 
        ax.set_yticks(np.arange(0, 0.8, 0.2))
    
    # Add light dashed horizontal grid lines and remove borders
    ax.grid(axis='y', linestyle='--', color='#cccccc', linewidth=0.3)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

# Adjust layout and save/show the figure
plt.tight_layout()
plt.savefig('comparison_plots_validation.pdf', format='pdf', dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```
